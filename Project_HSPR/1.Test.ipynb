```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
def get_student_discount_url():
    url = 'https://www.thsrc.com.tw'
    resp=requests.get(url + '/index.html')
    resp.encoding = 'utf-8'
    soup = bs(resp.text,'lxml')
    student_price_off = soup.find('a',{'title':'大學生優惠'})
    student_price_off_href = student_price_off['href']
    student_price_off_url = url + student_price_off_href
    resp2=requests.get(student_price_off_url)
    resp2.encoding = 'utf-8'
    soup2 = bs(resp2.text,'lxml')
    table_tag = soup2.find_all('section',{'class':'new_block'})
    table_url_set = table_tag[2].find_all('a')
    return table_url_set
```

```python

table_url = table_url_set[0]['href']
```

```python
resp3=requests.get(table_url)
resp3.encoding = 'utf-8'
soup3 = bs(resp3.text,'lxml')
time_table = soup3.find_all('div','timetable')
```

```python
southBound = time_table[0]
northBound = time_table[1]
```

```python
twentyPercentOff = '適用85折車次'
thirtyPercentOff = '適用7折車次'
fiftyPercentOff = '適用5折車次'
```

```python
southBoundRows = southBound.find_all('tr')
```

```python
southBoundBSTable = bs('','lxml')
for row in southBoundRows:
    cols = row.find_all('td')
    if(cols):
        for col in cols:
            if col:
                try:
                    if((int(col.text)) > 99 & (int(col.text)) < 10000):
                        southBoundBSTable.append(col.parent)
                except:
                    pass
southBoundTable = []
for x in southBoundBSTable.find_all('td'):
    element = []
    if(x.text == '—'):
        element.append(0)
    elif(x.text == '\xa0'):
        element.append(0)
    elif(x.text != ''):       
        element.append(x.text)        
    elif (x.find('img')['alt'] != None):
        if x.find('img')['alt'] == twentyPercentOff:
            element.append(1)
        elif  x.find('img')['alt'] == thirtyPercentOff:
            element.append(2)
        elif x.find('img')['alt'] == fiftyPercentOff:
            element.append(3)
        else:
            element.append('somethingWrong')
            print('somethingWrong')
    southBoundTable.append(element)
```

```python
count = 9
table = []
element = []
for x in southBoundTable:
    if(count == 0):
        table.append(element)
        count = 9
        element = []
        element.append(x)
        count -= 1
    else:
        element.append(x)
        count -= 1
table.append(element)
southBoundTable = table       
```
